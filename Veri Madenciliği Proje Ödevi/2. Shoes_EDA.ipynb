```python
dataset["Astar Materyali"] = dataset["Astar Materyali"].str.replace(" " , "0")
dataset["Astar Materyali"] = dataset["Astar Materyali"].str.replace('Tekstil', "1")
dataset["Astar Materyali"] = dataset["Astar Materyali"].str.replace('Suni Deri', "2")
dataset["Astar Materyali"] = dataset["Astar Materyali"].str.replace('Hakiki Deri', "3")
```

```python
#datasetteki NaN değer sayısının çok fazla olduğu gereksiz kolonlar atıldı. 200 den az olan değerler atıldı.
dataset = dataset.drop(["Ayakkabı Tabanı","Bilek Stili","Boy / Ölçü","Deri Kalitesi","Kalıp","Koleksiyon","Kumaş Teknolojisi",
                   "Okula Dönüş","Spor Branşı","Spor Türü","Sürdürülebilir","Sürdürülebilirlik Detayı","Taban","Zemin","Çeşit",
                   "Çivi Tipi","Özellik","Özellik","İzolasyon"],axis=1)
```

```python
plt.figure(figsize=(16,16))
plt.subplot(2,1,1)
sns.countplot(x='Renk',data= dataset,order = dataset['Renk'].value_counts().index)
plt.xticks(rotation = 90)
plt.show()
```

```python
for i in range(0,100):
    dataset["price"] = dataset.price.str.replace(',{}'.format(i), '')

dataset["price"] = dataset.price.str.replace(',00'.format(i), '')
dataset["price"] = dataset.price.str.replace(',01'.format(i), '')
dataset["price"] = dataset.price.str.replace(',02'.format(i), '')
dataset["price"] = dataset.price.str.replace(',03'.format(i), '')
dataset["price"] = dataset.price.str.replace(',04'.format(i), '')
dataset["price"] = dataset.price.str.replace(',05'.format(i), '')
dataset["price"] = dataset.price.str.replace(',06'.format(i), '')
dataset["price"] = dataset.price.str.replace(',07'.format(i), '')
dataset["price"] = dataset.price.str.replace(',08'.format(i), '')
dataset["price"] = dataset.price.str.replace(',09'.format(i), '')
dataset["price"] = dataset.price.str.replace('.', '')
```
